```python
data=pd.read_csv('Zomato_data.csv')
print(data)
```

```python
g0=sns.stripplot(x=data['listed_in(type)'],y=data['votes'])
```

```python
g=sns.stripplot(x=data['listed_in(type)'],y=data['votes'],jitter=True)
```

```python
g7=sns.stripplot(x='listed_in(type)',y='votes',jitter=True,hue='book_table',dodge=True,data=data)
```

```python
sns.set_style('white')
g1=sns.swarmplot(x=data['listed_in(type)'],y=data['votes'],s=1)
```

```python
sns.set_style('whitegrid')
g1=sns.swarmplot(y=data['listed_in(type)'],x=data['votes'],s=1.2)
```

```python
g9=sns.lineplot(x='listed_in(type)',y='rate',hue='book_table',data=data)
```

```python
g13=sns.lineplot(x='listed_in(type)',y='approx_cost(for two people)',data=data)
```

```python
g14=sns.lineplot(x='listed_in(type)',y='approx_cost(for two people)',hue='online_order',data=data)
```

```python
g3=sns.barplot(x='book_table',y='approx_cost(for two people)',data=data)
```

```python
g8=sns.barplot(x='book_table',y='votes',hue='listed_in(type)',data=data)
```
